```python
import matplotlib.pyplot as plt
import numpy as np
import random
import matplotlib

N = 1000
in_circle = 0
circle1 = plt.Circle((0, 0), 1, fill=False)
fig, ax = plt.subplots() # note we must use plt.subplots, not plt.subplot
plt.xlim(-1,1)
plt.ylim(-1.1)
ax.add_artist(circle1)

for i in range(N):
    x,y = 2*random.random()-1, 2*random.random()-1
    plt.scatter(x,y)


    if (np.sqrt(x**2 +y**2) < 1):
        in_circle += 1

4*in_circle/(N) # = pi
    
    
    
```
