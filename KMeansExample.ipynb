```python
# import modules
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# Generate Data Sets
dataset = make_blobs(n_samples=200, centers=4, n_features=2, cluster_std = 1.6, random_state=50)
```

```python
points = dataset[0]
```

```python
from sklearn.cluster import KMeans
```

```python
# Create kmeans object
kmeans = KMeans(n_clusters=4)

# fit kmeans object to dataset
kmeans.fit(points)
```

```python
plt.scatter(points[:,0], points[:,1])
```

```python
clusters = kmeans.cluster_centers_
```

```python
# print out clusters
print(clusters)
```

```python
# adjust position of centroids
y_km = kmeans.fit_predict(points)
```

```python
plt.scatter(points[y_km==0,0], points[y_km==0,1], s=50, color='red')
plt.scatter(points[y_km==1,0], points[y_km==1,1], s=50, color='green')
plt.scatter(points[y_km==2,0], points[y_km==2,1], s=50, color='yellow')
plt.scatter(points[y_km==3,0], points[y_km==3,1], s=50, color='cyan')

plt.scatter(clusters[0][0], clusters[0][1], marker='*', s=100, color='black')
plt.scatter(clusters[1][0], clusters[1][1], marker='*', s=100, color = 'black')
plt.scatter(clusters[2][0], clusters[2][1], marker='*', s=100, color='black')
plt.scatter(clusters[3][0], clusters[3][1], marker='*', s=100, color='black')

plt.show()
```
